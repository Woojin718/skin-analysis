```python
import keras
from keras.layers import Dense, BatchNormalization, Dropout, GlobalAveragePooling2D, Flatten, Conv2D, MaxPooling2D
from keras.preprocessing import image
from keras.regularizers import l2
from sklearn.model_selection import train_test_split
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import re
import glob
```

```python
# Parameters
BATCH_SIZE = 32
TARGET_SIZE = (224, 224)
INPUT_SHAPE= (224, 224, 3)
```

```python
densenet = tf.keras.applications.DenseNet169(
    include_top = False, 
    weights = "imagenet",
    input_shape = INPUT_SHAPE
    )

x = GlobalAveragePooling2D()(densenet.output)
x = Dense(units = 256, activation = 'relu')(x)
x = Dropout(0.4)(x)
x = BatchNormalization()(x)
x = Dense(units = 128, activation = 'relu')(x)
x = Dropout(0.1)(x)
x = Dense(units = 64, activation = 'relu')(x)
pred = Dense(units = 7, activation = 'softmax', kernel_regularizer = l2(l = 0.01))(x)
model = keras.models.Model(inputs = densenet.input, outputs = pred)
model.load_weights('./tmp/finalresult/densenet_skin_finetuned_best.h5')

x = model.layers[-8].output
x = Dense(256, activation='relu')(x)
x = BatchNormalization()(x)
x = Dense(128, activation='relu')(x)
x = BatchNormalization()(x)
x = Dense(64, activation='relu')(x)
x = Dense(32, activation='relu')(x)
output = Dense(7, activation='relu')(x)
model = keras.models.Model(inputs = model.input, outputs = output)
model.load_weights('./tmp/checkpoint(transferred_finetuning)/cp-0022.h5')
```

```python
model.compile(optimizer='adam', loss='mse', metrics=['mae','accuracy'])
```

```python
img = image.load_img("2000.jpg",target_size=(224,224,3))
img = image.img_to_array(img)
img = img/255
actual_score = [3, 3, 3, 4, 3, 3, 4]
```

```python
monitor = glob.glob("./monitor-images/*.jpg")
```

```python
score = []
for item in monitor:
    img = image.load_img(item,target_size=(224,224,3))
    img = image.img_to_array(img)
    img = img/255
    result = model.predict(img.reshape(1,224,224,3))[0]
    score.append([item,result])
    print(item,result)
```

```python
data = {
    "name":[item[0] for item in score],
    "texture":[item[1][0] for item in score],
    "pores":[item[1][1] for item in score],
    "spot":[item[1][2] for item in score],
    "saggy":[item[1][3] for item in score],
    "clear":[item[1][4] for item in score],
    "melanin":[item[1][5] for item in score],
    "wrinkles":[item[1][6] for item in score]
    }
```

```python
datatable = pd.DataFrame(data=data)
```
